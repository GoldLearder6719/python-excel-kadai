```python
from openpyxl import Workbook
from datetime import datetime

# 現在日付を取得
today = datetime.now()
today_str = today.strftime("%Y/%m/%d")      # 表示用
file_date = today.strftime("%Y%m%d")        # ファイル名用

# Workbook作成
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# ------------------------
# タイトル
# ------------------------
ws["B2"] = "請求書"

# ------------------------
# 固定データ
# ------------------------
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# ------------------------
# 請求情報（修正：日付を自動生成）
# ------------------------
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today_str

# ------------------------
# テーブルヘッダ
# ------------------------
ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

# ------------------------
# 商品データ
# ------------------------
ws["B11"] = "商品A"
ws["C11"] = 2
ws["D11"] = 10000
ws["E11"] = 20000

ws["B12"] = "商品B"
ws["C12"] = 1
ws["D12"] = 15000
ws["E12"] = 15000

# ------------------------
# 合計（修正あり）
# ------------------------
ws["E13"] = "=SUM(E11:E12)"

ws["B15"] = "小計"
ws["E15"] = "=E13"

ws["B16"] = "消費税"
ws["E16"] = "=E15*0.1"

ws["B17"] = "合計"
ws["E17"] = "=E15+E16"

# ------------------------
# 保存（修正：ファイル名）
# ------------------------
file_name = f"請求書_{file_date}.xlsx"
wb.save(file_name)
```
